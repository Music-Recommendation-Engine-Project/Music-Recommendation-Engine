Get list of song_id to track_uri, several songs have multiple track_uri

```python
# import libraries
import pandas as pd
import polars as pl
```

```python
# Read the data 
df_user_songMD = pl.read_parquet('user_songMD_duplicates.parquet')
df_user_songs_IDs = df_user_songMD[['song_id', 'track_uri']]
df_user_songs_IDs = df_user_songs_IDs.unique()
# group by song_id witht the multiple song_id
df_user_songs_IDs = df_user_songs_IDs.groupby('song_id').agg(pl.col('track_uri').alias('track_uri'))
```

```python
df_user_songs_IDs
```

```python
# get only the song_id that have more than one track_uri
for i in range(len(df_user_songs_IDs)):
    if len(df_user_songs_IDs['track_uri'][i]) > 50:
        print(df_user_songs_IDs['song_id'][i])
        print(df_user_songs_IDs['track_uri'][i])
```

```python
# Save the data
df_user_songs_IDs.write_parquet('user_songs_IDs.parquet')
```
