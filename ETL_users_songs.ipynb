```python
df_songsMD = pl.read_csv('input_data/tracks.csv')
# Rename columns in polars
df_songsMD = df_songsMD.rename({'id':'track_uri', 'name':'track_name'})
```

Transformation of columns artists and id_artists to select the first artist. This is for standardizing the merging of the data 

```python
df_songsMD = df_songsMD.to_pandas()
# Transform the df[artists] column
df_songsMD['artists'] = df_songsMD['artists'].apply(lambda x: x.replace('[', '').replace(']', '').replace("'", ""))
# Take the first artist
df_songsMD['artist_name'] = df_songsMD['artists'].apply(lambda x: x.split(',')[0])
# Drop the artists column
df_songsMD.drop('artists', axis=1, inplace=True)

# Transform the df[id_artist] column
df_songsMD['id_artists'] = df_songsMD['id_artists'].apply(lambda x: x.replace('[', '').replace(']', '').replace("'", ""))
# Take the first artist
df_songsMD['artist_id'] = df_songsMD['id_artists'].apply(lambda x: x.split(',')[0])
# Drop the id_artist column
df_songsMD.drop('id_artists', axis=1, inplace=True)
# df pandas to polars
df_songsMD = pl.from_pandas(df_songsMD)
```
